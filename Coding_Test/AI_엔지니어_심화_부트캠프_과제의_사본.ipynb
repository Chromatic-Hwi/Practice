````python
# 답안 작성

class PhoneBook:
    def __init__(self):
        self.contacts = {}

    def add_contact(self, name, phone_number):
        self.contacts[name] = phone_number

class PhoneBookFunc(PhoneBook):
    def __init__(self):
        super().__init__()

    def add_num1(self, name, phone_number):
        if name in self.contacts:
            print(f"\n[추가불가] >> {name}의 연락처가 이미 존재합니다.\n")
        else:
            PhoneBook.add_contact(self, name, phone_number)
            print(f"\n[추가완료] >> {name}의 연락처: {self.contacts[name]}\n")

    def update_num2(self, name, phone_number):
        if name in self.contacts:
            self.contacts[name] = phone_number
            print(f"\n[수정완료] >> {name}의 연락처: {self.contacts[name]}\n")
        else:
            print(f"\n[수정불가] >> {name}의 연락처가 존재하지 않습니다.\n")

    def search_num3(self, name):
        if name in self.contacts:
            print(f"\n[검색완료] >> {name}: {self.contacts[name]}\n")
        else:
            print(f"\n[검색완료] >> {name}의 연락처가 존재하지 않습니다.\n")

    def delete_num4(self, name):
        if name in self.contacts:
            del self.contacts[name]
            print(f"\n[삭제완료] {name}의 연락처가 삭제되었습니다.\n")
        else:
            print(f"\n[삭제불가] >> {name}의 연락처가 존재하지 않습니다.\n")

    def clear_num5(self):
        yn = input("정말 초기화 하시겠습니까? (Y/N)").upper()
        if yn == "Y":
            self.contacts.clear()
            print("\n[초기화완료] >> 모든 연락처가 초기화되었습니다.\n")
        else:
            print("\n[초기화취소] >> 초기화를 취소합니다.\n")
````

````python
def print_menu():
    print("\n<< 연락처 관리 메뉴 >>")
    print("1. 연락처 추가")
    print("2. 연락처 수정")
    print("3. 연락처 조회")
    print("4. 연락처 삭제")
    print("5. 모든 연락처 초기화")
    print("6. 인터페이스 종료")

def main():
    phonebook = PhoneBookFunc()
    while True:
        print_menu()
        print()
        choice = input("메뉴를 선택해주세요: ")
        try:
            if choice == '1':
                print("\n※ 연락처 관리 메뉴 1번 [추가] 입니다.")
                name = input("이름을 입력해주세요.:")
                phone_number = input("연락처를 입력해주세요.:")
                #phonebook.add_contact(name, phone_number)
                phonebook.add_num1(name, phone_number)

            elif choice == '2':
                print("\n※ 연락처 관리 메뉴 2번 [수정] 입니다.")
                name = input("이름을 입력해주세요.:")
                phone_number = input("연락처를 입력해주세요.:")
                phonebook.update_num2(name, phone_number)

            elif choice == '3':
                print("\n※ 연락처 관리 메뉴 3번 [조회] 입니다.")
                name = input("이름을 입력해주세요.:")
                phonebook.search_num3(name)

            elif choice == '4':
                print("\n※ 연락처 관리 메뉴 4번 [삭제] 입니다.")
                name = input("이름을 입력해주세요.:")
                phonebook.delete_num4(name)

            elif choice == '5':
                print("\n※ 연락처 관리 메뉴 5번 [초기화] 입니다.")
                phonebook.clear_num5()

            elif choice == "6":
              print("연락처 관리 서비스를 종료합니다.")
              break
            else:
                print("※ 없는 메뉴입니다! 다시 시도해주세요.\n")

        except ValueError as e:
            print(f"Error: {e}")
````

````python
# 답안 작성

test_nums = range(2, 121)
answer_nums = []

for num in test_nums:
    num_root = num ** 0.5
    for i in range(2, int(num_root) + 1):
        if num % i == 0:
            break
    else:
        answer_nums.append(num)

print(answer_nums)
````
